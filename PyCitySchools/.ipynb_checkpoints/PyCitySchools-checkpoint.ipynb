# DISTRICT SUMMARY

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_merged = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_merged.head(20)
```

```python
# Create array of unique school names
unique_school_names = school_data_merged["school_name"].unique()

# Calculate the number of schools
Total_Schools = len(unique_school_names)

# Calculate the total students
Total_Students = school_data_merged["student_name"].count()

# Calculate Total Budget Allocation
Total_Budget = school_data["budget"].sum()

# Calculate average Math Score
Average_Math_Score = school_data_merged["math_score"].mean()

# Calculate average Reading Score
Average_Reading_Score = school_data_merged["reading_score"].mean()

# Calculate % Passing Math
num_passing_math = student_data.loc[student_data['math_score'] >= 70]['math_score'].count()
pct_pass_math = num_passing_math/Total_Students

# Calculate % Passing Reading 
num_passing_reading = student_data.loc[student_data['reading_score'] >= 70]['reading_score'].count()
pct_pass_reading = num_passing_reading/Total_Students  

# Calculate % Overall passing
overall_pass = student_data[(student_data['math_score'] >= 70) & (student_data['reading_score'] >= 70)].count()['student_name']                        
pct_overall_pass = overall_pass/Total_Students                                       

# Place all the Data into a summary dataframe
district_summary_df = pd.DataFrame({"Total Schools" : [Total_Schools],
                                    "Total Students" : [Total_Students],
                                    "Total Budget" : [Total_Budget],
                                    "Average Math Score" : [Average_Math_Score],
                                    "Average Reading Score" : [Average_Reading_Score],
                                    "% Passing Math": [pct_pass_math],
                                    "% Passing Reading": [pct_pass_reading],
                                    "% Overall Passing": [pct_overall_pass]})

district_summary_column = district_summary_df[["Total Schools",
                                               "Total Students",
                                               "Total Budget",
                                               "Average Reading Score",
                                               "% Passing Math",
                                               "% Passing Reading",
                                               "% Overall Passing"]]
                                       
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)  
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format) 
district_summary_df["% Passing Math"] = (district_summary_df["% Passing Math"]*100).map("{:.8}".format)
district_summary_df["% Passing Reading"] = (district_summary_df["% Passing Reading"]*100).map("{:.8}".format)
district_summary_df["% Overall Passing"] = (district_summary_df["% Overall Passing"]*100).map("{:.8}".format)

district_summary_df.head()                              
```
